```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3=np.percentile(col,[25,75])
    IQR=Q3-Q1
    lower_range= Q1-(1.5 * IQR)
    upper_range= Q3+(1.5 * IQR)
    return lower_range, upper_range
```

```python
for i in range(len(Cont_Col)):
    lr, ur=remove_outlier(df_sales[[Cont_Col[i]]])
    df_sales[[Cont_Col[i]]]=np.where(df_sales[[Cont_Col[i]]]<lr,lr,df_sales[[Cont_Col[i]]])
    df_sales[[Cont_Col[i]]]=np.where(df_sales[[Cont_Col[i]]]>ur,ur,df_sales[[Cont_Col[i]]])
```
